```python
# mengimport library pandas
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as s
import math
import seaborn as sns


# membaca file csv
df = pd.read_csv("water_potability.csv",
        names = ["id", "pH", "Hardness", "Solids", "Chloramines", "Sulfate", "Conductivity", "OrganicCarbon", "Trihalomethanes",
                "Turbidity", "Potability"])
```

```python
#With shapiro wilks
# Untuk meriksa aja
names = ["pH", "Hardness", "Solids", "Chloramines", "Sulfate", "Conductivity", "OrganicCarbon", "Trihalomethanes",
                "Turbidity" ]

for x in names:
    newDf= pd.DataFrame(data = df, columns = [x])
    stat, p = s.shapiro(newDf) 
    print(x , end=" ")
    if p > 0.05:
        print("Distribusi Normal")
    else:
        print("Tidak normal")
```
